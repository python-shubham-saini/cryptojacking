```python
import pandas as pd
import numpy as np
df1 = pd.read_csv("final-dataset.csv")
df1.head()
```

```python
#get all columns
# sorted(df)
df1 = df1.select_dtypes(exclude=['object'])
df1
```

```python
col = sorted(df1)
d = []
for i in col:
    c = df1[i]
    if len(c.unique())==1 :
        d.append(i)
# print(d)
df1 = df1.drop(d, axis='columns')
# sorted(df1)
```

```python
df1 = df1[~df1.isin([np.nan, np.inf, -np.inf]).any(1)]
df1['Anormal'] = df1['Anormal'].astype(int)
df1.info()
```

```python
#Split the dataset into feature and target set

X = df1.copy()
X.drop(list(X.filter(regex='mem_')), axis = 1, inplace = True)
X.drop(list(X.filter(regex = 'memswap')), axis = 1, inplace = True)
X = X.iloc[:, :-1]
y = df1['Anormal'].astype(int)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
dtc.score(X_test, y_test)
```

```python
y_predict = dtc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_test,y_predict)
cm
```

```python
cr = classification_report(y_test, y_predict)
# cr
print(classification_report(y_test, y_predict))
```
